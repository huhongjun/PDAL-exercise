```python
# 针对煜邦电力，一段电力通道点云分好类之后，按地物分类拆解成数字文件，文件数量不固定，因为某些区段可能没有某个类别
# 此处去掉了地面点【2】，手动写好待处理文件列表与分类列表

ccode = [6, 16, 5, 14, 
         3, 4, 8, 21, 23, 
         12, 19, 18, 20, 17,
        15]
cname = ['buildings','conductor','high_vegetation','lake',
         'low_vegetation','mid_vegetation','mkp_ground','other','other_line',
         'road','scissors_crossing(down)','scissors_crossing(up)','shield_wire','structures',
         'substations']
```

# las转换csv格式,按文件名赋值分类

```python
# 按文件列表，生成PDAL转换器需要的json字符串变量

import pdal

# 待处理电力走廊工作路径，其下有input、output子目录
folder = "workshop\\cz_1-9"

for i in range(len(ccode)) :
    filename = folder +'_'+ cname[i]
    classification = ccode[i]
    
    json =  """
    {  "pipeline": [
      { "type" : "readers.las", "filename" : "%s\\input\\%s.las"      },
      { "type" : "filters.assign",  "assignment" : "Classification[0:32]=%s"      },
      { "type" : "writers.text", "format": "csv", "filename" : "%s\\output\\%s.csv"      }  ]}
      """ % (folder,filename,classification,folder,filename)
    
    print(json)
    print(" - Code: %2d, Name: %35s.las, Count: %6d " % (classification,filename,0))
```

# las转换las格式,按文件名赋值分类

```python
import pdal

folder = "cz_1-9"
for i in range(len(ccode)) :
    filename = folder +'_'+ cname[i]
    classification = ccode[i]
    
    json =  """
    {  "pipeline": [
      { "type" : "readers.las", "filename" : "workshop\\\\%s\\\\input\\\\%s.las"      },
      { "type" : "filters.assign",  "assignment" : "Classification[0:32]=%s"      },
      { "type" : "writers.las", "filename" : "workshop\\\\%s\\\\output\\\\%s.las"      }  ]}
      """ % (folder,filename,classification,folder,filename)
    print(json)
    
    pipeline = pdal.Pipeline(json)
    pipeline.validate()        # check if our JSON and options were good
    pipeline.loglevel = 8      #really noisy
    count = pipeline.execute()    

    print(" - Code: %2d, Name: %35s.las, Count: %6d " % (classification,filename,count))
```

```python
# 生成文件列表字符串

folder = "cz_1-9"

filelist=''
for i in range(len(ccode)) :
    filename = '"workshop/%s/input/%s_%s.las"' % (folder,folder,cname[i])
    filelist = filelist  + filename + ','
print(filelist)
```

```python
# las文件合并

json_merge = """{
  "pipeline": [%s
    {
      "type": "filters.merge"
    },
    "workshop/cz_1-9/output/cz_1-9_merge.las"
  ]
}
""" % filelist

print(json_merge)
```

```python
# las文件分拆，按长度大小 =》有些拆出来的文件很小，并不是平均比例

json_splitter = """
{
    "pipeline":
        [
            "workshop\\\\cz_1-9\\\\output\\\\cz_1-9_merge.las",
            {
                "type":"filters.splitter",
                "length":"2000"
            },
            {
                "type":"writers.las",
                "filename":"workshop\\\\cz_1-9\\\\output\\\\cz_1-9_merge_splitter_#.las"
            }
        ]
}
"""
```

```python
# las文件分拆，按分拆文件数量

json_divide = """
{
    "pipeline":
        [
            "workshop\\\\cz_1-9\\\\output\\\\cz_1-9_merge.las",
            {
                "type":"filters.divider",
                "count":"3"
            },
            {
                "type":"writers.las",
                "filename":"workshop\\\\cz_1-9\\\\output\\\\cz_1-9_merge_divide_#.las"
            }
        ]
}
"""
```

```python
# las切片，可以设定文件包含点数量的最大值

json_chipper = """
{
    "pipeline":
        [
            "workshop\\\\cz_1-9\\\\output\\\\cz_1-9_merge.las",
            {
                "type":"filters.chipper",
                "capacity":"8000000"
            },
            {
                "type":"writers.las",
                "filename":"workshop\\\\cz_1-9\\\\output\\\\cz_1-9_merge_chipper_#.las"
            }
        ]
}
"""
```

```python
# 按三维网格下采样，分类和RGB信息会丢失

json_voxelgrid = """
{
    "pipeline":
        [
            "workshop\\\\cz_1-9\\\\output\\\\cz_1-9_merge_chipper_3.las",
            {
                "type":"filters.voxelgrid",
                "leaf_z":"0.4"
            },
            {
                "type":"writers.las",
                "filename":"workshop\\\\cz_1-9\\\\output\\\\cz_1-9_merge_chipper_3_voxelgrid_04.las"
            }
        ]
}
"""
```
